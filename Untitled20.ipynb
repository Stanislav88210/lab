```python
# параметры вина, будем смотреть зависимости параметров в вине
useful_cols = ['Alcohol', 'Malic acid', 'Alcalinity of ash', 
               'Flavanoids' 
              ]
df[useful_cols].head(10)
```

```python
#параметры алкоголя в вине и пр
df[useful_cols].plot();
```

```python
df[useful_cols].plot(kind='bar', rot=45);
```

```python
df[useful_cols][list(filter(lambda x: x != 'Global_Sales', df[useful_cols].columns))]\
    .plot(kind='bar', rot=45, stacked=True);
```

```python
df[useful_cols][list(filter(lambda x: x != 'Global_Sales', df[useful_cols].columns))]\
    .plot(kind='area', rot=45, stacked=True);
```

```python

%config InlineBackend.figure_format = 'png' 
sns_plot = sns.pairplot(
    df[['Alcohol', 'Malic acid', 'Flavanoids']]);
sns_plot.savefig('pairplot.png')
```
