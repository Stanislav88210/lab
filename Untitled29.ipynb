```python
print("Исходные признаки:\n", list(raw_df.columns), "\n") 
data_dummies = pd.get_dummies(raw_df)
data_dummies=data_dummies.drop(['SaleCategory'],1)
print("Признаки после get_dummies:\n", list(data_dummies.columns))
```

```python
features = data_dummies.ix[:, 'Condition1_Artery':'SaleCondition_Partial']
# Извлекаем массивы NumPy
X = features.values
y = raw_df['SaleCategory'].values
print("форма массива X: {} форма массива y: {}".format(X.shape, y.shape))
```

```python
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(features, raw_df[['SaleCategory']], test_size = 0.3, random_state = 0)
model = linear_model.SGDClassifier(alpha=0.0001, n_iter=100, random_state = 0)
model.fit(train_data, train_labels)
model_predictions = model.predict(test_data)
print (metrics.accuracy_score(test_labels, model_predictions))
print (metrics.classification_report(test_labels, model_predictions))
```
